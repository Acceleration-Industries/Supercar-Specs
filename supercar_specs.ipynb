```python
#1. Class vs. Object:
#Class: Defines a blueprint, such as Supercar, Ferrari488Pista2020, and PorscheGT2RS2018 in our code.
#Object: Represents instances created from classes, like ferrari_488_pista_2020 and porsche_gt2_rs_2018.

#2. Encapsulation:
#Encapsulation is applied in the Supercar class.
#Attributes and getter methods like get_model and get_horsepower are used for controlled access to data.

#3. Inheritance:
#Inheritance is utilized through PorscheGT2RS2018 and Ferrari488Pista2020 classes inheriting from Supercar.
#Allows subclasses to have unique features and override methods while retaining common attributes.

#4. Polymorphism:
#Polymorphism is demonstrated in the supercar_specs function.
#Common interface enables interaction with different supercars using the same method names like get_top_speed.

#5. Abstraction:
#Abstraction is achieved with the Supercar class.
#Represents a basic supercar, keeping detailed information in subclasses (PorscheGT2RS2018 and Ferrari488Pista2020).

from IPython.display import Image, display
class Supercar:
    def __init__(drift, model, horsepower, torque, top_speed, weight, acceleration, engine_type):
        drift._model = model
        drift._horsepower = horsepower
        drift._torque = torque
        drift._top_speed = top_speed
        drift._weight = weight
        drift._acceleration = acceleration
        drift._engine_type = engine_type

    def get_model(drift):
        return drift._model

    def get_horsepower(drift):
        return drift._horsepower

    def get_torque(drift):
        return drift._torque

    def get_top_speed(drift):
        return drift._top_speed

    def get_weight(drift):
        return drift._weight

    def get_acceleration(drift):
        return drift._acceleration

    def get_engine_type(drift):
        return drift._engine_type

    def display_image(drift, image_filename):
        display(Image(filename=image_filename))

    def get_display_info(drift):
        return f"Model: {drift.get_model()}, Horsepower: {drift.get_horsepower()} hp, " \
               f"Torque: {drift.get_torque()} lb-ft, Top Speed: {drift.get_top_speed()} mph, " \
               f"Curb Weight: {drift.get_weight()} lbs, 0-60 Time: {drift.get_acceleration()} seconds, " \
               f"Engine Type: {drift.get_engine_type()}"


class Ferrari488Pista2020(Supercar):
    def __init__(drift):
        super().__init__(model="Ferrari 488 Pista", horsepower=None, torque=None, top_speed=None, weight=None, acceleration=None, engine_type=None)

    def set_specifications(drift):
        drift._horsepower = 710
        drift._torque = 568
        drift._top_speed = 211
        drift._weight = 3054
        drift._acceleration = 2.85
        drift._engine_type = "Twin-Turbo 3.9L V8"


class PorscheGT2RS2018(Supercar):
    def __init__(drift):
        super().__init__(model="Porsche 911 GT2RS", horsepower=None, torque=None, top_speed=None, weight=None, acceleration=None, engine_type=None)

    def set_specifications(drift):
        drift._horsepower = 700
        drift._torque = 553
        drift._top_speed = 211
        drift._weight = 3241
        drift._acceleration = 2.7
        drift._engine_type = "Twin-Turbo 3.8-liter Flat-Six"


def supercar_specs():
    display(Image(filename='ferrari_porsche.jpg'))

    while True:
        car_choice = input("Choose a Supercar - Ferrari or Porsche (Type 'EXIT' to Exit): ").lower()

        if car_choice == 'exit':
            print("Thank You for Visiting Supercar Specs!")
            break

        car_driving = None

        if car_choice == "ferrari":
            ferrari_488_pista_2020 = Ferrari488Pista2020()
            ferrari_488_pista_2020.set_specifications()
            ferrari_488_pista_2020.display_image('ferrari.jpg')
            car_driving = ferrari_488_pista_2020
        elif car_choice == "porsche":
            porsche_gt2_rs_2018 = PorscheGT2RS2018()
            porsche_gt2_rs_2018.set_specifications()
            porsche_gt2_rs_2018.display_image('porsche.jpg')
            car_driving = porsche_gt2_rs_2018
        else:
            print("Invalid Choice. Enter 'Ferrari' or 'Porsche'.")
            continue

        print(f"You Chose the {car_driving.get_model()}!")

        while True:
            print("\nChoose a Specification:")
            print("1. Top Speed")
            print("2. Horsepower")
            print("3. Torque")
            print("4. 0-to-60 mph")
            print("5. Curb Weight")
            print("6. Engine Type")
            print("Enter 'EXIT' to Go Back.")

            specification_choice = input("Enter Your Choice: ").lower()

            if specification_choice == 'exit':
                break

            if specification_choice in ['1', 'top speed']:
                print(f"The Top Speed for a {car_driving.get_model()} is: {car_driving.get_top_speed()} MPH")
            elif specification_choice in ['2', 'horsepower']:
                print(f"The Horsepower for a {car_driving.get_model()} is: {car_driving.get_horsepower()} HP")
            elif specification_choice in ['3', 'torque']:
                print(f"The Torque for a {car_driving.get_model()} is: {car_driving.get_torque()} lb-ft")
            elif specification_choice in ['4', '0-to-60 mph']:
                print(f"The 0-to-60 Time for a {car_driving.get_model()} is: {car_driving.get_acceleration()} Seconds")
            elif specification_choice in ['5', 'curb weight']:
                print(f"The Curb Weight for a {car_driving.get_model()} is: {car_driving.get_weight()} Lbs.")
            elif specification_choice in ['6', 'engine type']:
                print(f"The Engine Type for a {car_driving.get_model()} is: {car_driving.get_engine_type()}")
            else:
                print("Invalid Choice. Choose a Specification or Enter 'EXIT'.")

supercar_specs()
```
